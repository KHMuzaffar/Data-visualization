```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df_fuel_engine_type=pd.read_csv("stacked_area_chart_data.csv")
df_fuel_engine_type
```

```python
colors=["#011638","#7e2987","#ef2026"]
labels=["Diesel","Petrol","Gas"]
plt.figure(figsize=(12,6))
plt.stackplot(df_fuel_engine_type["Year"],
              df_fuel_engine_type["Diesel"],
              df_fuel_engine_type["Petrol"],
              df_fuel_engine_type["Gas"],
              colors=colors,
              edgecolor='none')
plt.xticks(df_fuel_engine_type["Year"],rotation=45) 
plt.legend(labels=labels,loc="upper left")
plt.ylabel("Number of Cars",fontsize=13)
plt.title("Popularity of Engine Fuel Types (1982-2016)",fontsize=14,weight="bold")
sns.despine()
plt.show()
plt.savefig("Fuel types Stacked_area.png")
```
